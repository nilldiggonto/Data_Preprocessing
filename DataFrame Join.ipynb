```python
####################
# Colulmn labels and index joining
###################
```

```python
data2004.columns = ['Name','Medals']
```

```python
data2004.head()
```

```python
data2008.head()
```

```python
new_merge = data2004.merge(data2008,how='outer',left_on='Name', right_on='Athlete',
                          suffixes=['_2004','_2008'],indicator=True)
```

```python
new_merge
```

```python
new_merge['Name'].fillna(new_merge['Athlete'],inplace=True)
```

```python
new_merge
```

```python
new_merge.drop(['Athlete','_merge'],axis=1,inplace=True)
```
